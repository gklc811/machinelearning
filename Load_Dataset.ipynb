```python
import sklearn.datasets
import numpy as np

# change dataset here
dataset = sklearn.datasets.load_breast_cancer()
print (dataset)
X = dataset.data
Y = dataset.target
X_header = dataset.feature_names
Y_header = dataset.target_names

# Standard sklearn dataset format

print(X.shape)
print(Y.shape)
print(X_header)
print(Y_header)
```

```python
import pandas as pd

# pd.set_option('display.max_columns', None)

# Convert sklearn dataset tp pandas dataframe

df = pd.DataFrame(X, columns=X_header)
df['class'] = Y

df.head()
```

```python
# get high level statistics
df.describe()
```

```python
df.head()
```

```python
groups = df.groupby(['class']).groups
print(groups)
```

```python
g_df = df.groupby(['class'])

for name,group in g_df:
    print (name)
    print (group.head())
    print("************************************")
```

```python
# Getting grouped df
g_df.get_group(1)
```
