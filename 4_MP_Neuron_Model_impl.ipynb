```python
import sklearn.datasets
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import accuracy_score

dataset = sklearn.datasets.load_breast_cancer()
# print(dataset)
X = dataset.data
Y = dataset.target
X_header = dataset.feature_names
Y_header = dataset.target_names

df = pd.DataFrame(X, columns=X_header)
df['class'] = Y

df_y = df['class']
df_x = df.drop(columns=['class'])

X_train, X_test, Y_train, Y_test = train_test_split(df_x, df_y, test_size=0.1, stratify = df_y, random_state =1)
x_binarirised_train = X_train.apply(pd.cut, bins=2,labels=[1,0])
x_binarirised_test = X_test.apply(pd.cut, bins=2,labels=[1,0])

x_binarirised_train.head()
```

```python
b = 1;
for x in x_binarirised_train.columns:
    analyzer_df = pd.DataFrame()
    analyzer_df['sum'] = x_binarirised_train.sum(axis=1)
    analyzer_df['pred'] = (analyzer_df['sum'] >= b) * 1
    analyzer_df['class'] = Y_train
    analyzer_df['accurate'] = ((analyzer_df['class'] ==  analyzer_df['pred']) ) *1
    correct_predictions = analyzer_df.sum()['accurate']
    total_predictions = x_binarirised_train.count()[0]
    accuracy = correct_predictions/total_predictions
    print(b, accuracy)
    b +=1
```

```python
# Numpy version

x_test_vals = x_binarirised_train.values


for b in range(x_test_vals.shape[1]):
    b += 1
    y_pred = []
    for x, y  in zip(x_test_vals,Y_train):
        y_pred.append((x.sum()>= b)*1)
    accuracy = accuracy_score(Y_train, y_pred)
    print(b, accuracy)
```

```python
# MP Neuron Class
class MPNeuron:
    def __init__(self):
        self.b = None
    
    def model(self,x):
        return ((sum(x) >= self.b) * 1)
    
    def predict(self,X):
        Y=[]
        for x in X:
            Y.append(self.model(x))
        return Y
    
    def fit(self,X,Y):
        accuracy ={}
        for b in range(X.shape[1]):
            self.b = b + 1
            y_pred = self.predict(X)
            accuracy[self.b] = accuracy_score(Y, y_pred)
#             print( self.b, accuracy[self.b])
        self.b = max(accuracy, key=accuracy.get)
```

```python
mp_n = MPNeuron()
x_test_vals = x_binarirised_train.values
mp_n.fit(x_test_vals, Y_train)
x_test_vals = x_binarirised_test.values
y_pred = mp_n.predict(x_test_vals)
print(accuracy_score(y_pred, Y_test))
```
