```python
import sklearn.datasets
import pandas as pd
from sklearn.model_selection import train_test_split

dataset = sklearn.datasets.load_breast_cancer()
# print(dataset)
X = dataset.data
Y = dataset.target
X_header = dataset.feature_names
Y_header = dataset.target_names

df = pd.DataFrame(X, columns=X_header)
df['class'] = Y

df_y = df['class']
df_x = df.drop(columns=['class'])

X_train, X_test, Y_train, Y_test = train_test_split(df_x, df_y, test_size=0.8, stratify = df_y, random_state =1)
print(df_y.mean(), Y_test.mean(),Y_train.mean())
```

```python
import matplotlib.pyplot as plt
plt.plot(X_train)
plt.show()
```
